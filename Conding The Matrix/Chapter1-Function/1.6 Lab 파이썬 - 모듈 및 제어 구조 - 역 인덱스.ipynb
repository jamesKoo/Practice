```python
def makeInverseIndex(strlist): 
    inv_idx = {}
    for (i, doc) in enumerate(strlist):
        for word in doc.split():
            if word in inv_idx:
                inv_idx[word].add(i)
            else:
                inv_idx[word] = set()
                inv_idx[word].add(i)            
    return inv_idx

f = open('stories_small.txt')
strlist = list(f)
inv_idx = makeInverseIndex(strlist)
inv_idx
```
