```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2

def imshow(img1, img2):
    fig, axis = plt.subplots(ncols=2, figsize=(10, 10))
    
    if len(img1.shape) == 3:
        axis[0].imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
        axis[1].imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
    else:
        axis[0].imshow(img1, cmap='gray')
        axis[1].imshow(img2, cmap='gray')
    plt.show()
   
    
# Read the images to be aligned
img1 = cv2.imread('Resource/imageAlign1.jpg')
img2 = cv2.imread('Resource/imageAlign2.jpg')

# Convert images to graysclae
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

imshow(img1, img2)
```

```python
# Find size of image1
size = img1.shape

# Define the motion model
warp_mode = cv2.MOTION_EUCLIDEAN

# Define 2x3 or 3x3 matrices and initialize the matrix to identify
if warp_mode == cv2.MOTION_HOMOGRAPHY:
    warp_matrix = np.eye(3, 3, dtype=np.float32)
else:
    warp_matrix = np.eye(2, 3, dtype=np.float32)

# Specify the number of iterations
n_iter = 5000

# Specify the threshold of the increment
# in the correlation coefficient between two iterations
term_eps = 1e-10

# Define termination criteria
criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, n_iter, term_eps)

# Run the ECC algorithm. The results are stored in wrap_matrix.
(cc, warp_matrix) = cv2.findTransformECC(img1_gray, img2_gray, warp_matrix, warp_mode, criteria)
print(warp_matrix)
```

```python
if warp_mode == cv2.MOTION_HOMOGRAPHY:
    img2_aligned = cv2.warpPerspective(img2, warp_matrix, (size[1], size[0]),
                                       flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP)
else:
    img2_aligned = cv2.warpAffine(img2, warp_matrix, (size[1], size[0]),
                                       flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP)
imshow(img1, img2_aligned)
```

### Approximation of the image gradient
The three channels of the image are more strongly correlated in the gradient domin.

```python
def get_gradient(img):
    # Calculate the x and y gradients using Sobel operator
    grad_x = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=3)
    grad_y = cv2.Sobel(img, cv2.CV_32F, 0, 1, ksize=3)
    
    # Combine the two gradients
    grad = cv2.addWeighted(np.absolute(grad_x), 0.5, np.absolute(grad_y), 0.5, 0)
    
    return grad
```

```python
# Read 8-bit color iimage.
# This is an image in which the three channels are concateated vertically.
img = cv2.imread('Resource/emir.jpg', cv2.IMREAD_GRAYSCALE)

# Find the width and height of the color image
size = img.shape
height = int(size[0] // 3)
width = size[1]

# Extract the three channels from the gray scale image 
# and merge the three Channels into one color image
img_color = np.zeros((height, width, 3), dtype=np.uint8)
for i in range(3):
    img_color[:,:,i] = img[i*height:(i+1)*height]

plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
```

```python
# Allocate space for aligned image
img_aligned = np.zeros((height, width, 3), dtype=np.uint8)

# The blue and green channels will be aligned to the red chnnnel.
# So copy the red channel
img_aligned[:,:,2] = img_color[:,:,2]

# Define motion model
warp_mode = cv2.MOTION_HOMOGRAPHY
 
# Set the warp matrix to identity.
warp_matrix = np.eye(3, 3, dtype=np.float32)

# Set the stopping criteria for the algorithm.
criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, n_iter, term_eps)

# Wrap the blue and green channels to the red channel
for i in range(2):
    (cc, wrap_matrix) = cv2.findTransformECC(get_gradient(img_color[:,:,2]), get_gradient(img_color[:,:,i]),
                                                         warp_matrix, warp_mode, criteria)
    img_aligned[:,:,i] = cv2.warpPerspective(img_color[:,:,i], warp_matrix, (width, height),
                                            flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP)
    print(i, wrap_matrix) 
```

```python
imshow(img_color, img_aligned)
```
