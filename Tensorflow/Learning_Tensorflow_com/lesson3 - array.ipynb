```python
#1 

import numpy as np
import tensorflow as tf
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# First, load the image again
filename = 'MarshOrchid.jpg'
image = mpimg.imread(filename)
height, width, depth = image.shape

# Create a Tensorflow Variable
x = tf.Variable(image, name='x')

model = tf.global_variables_initializer()

with tf.Session() as sess:
    x = tf.reverse_sequence(x, [width]*height, 1, batch_dim=0)
    x = tf.transpose(x, perm=[1, 0, 2])
    sess.run(model)
    result = sess.run(x)
    
print(result.shape)
plt.imshow(result)
plt.show()
```

```python
#2

import numpy as np
import tensorflow as tf
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# First, load the image again
filename = 'MarshOrchid.jpg'
image = mpimg.imread(filename)

# Create a Tensorflow Variable
x = tf.Variable(image, name='x')

model = tf.global_variables_initializer()

with tf.Session() as sess:
    shape = tf.shape(x)
    sess.run(model)
    shape = sess.run(shape)
    
    x = tf.reverse_sequence(x, np.ones(shape[0])*shape[1], 1, batch_dim=0)
    result = sess.run(x)    
    
print(result.shape)
plt.imshow(result)
plt.show()
```

```python
#5

import numpy as np
import tensorflow as tf
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# First, load the image again
filename = 'MarshOrchid.jpg'
image = mpimg.imread(filename)
height, width, depth = image.shape

# Create a Tensorflow Variable
x = tf.Variable(image, name='x')
mirror_mask = np.ones(height)*(width/2)

model = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(model)
    
    x = tf.reverse_sequence(x, mirror_mask, 1, batch_dim=0)
    result = sess.run(x)    
    
print(result.shape)
plt.imshow(result)
plt.show()
```
